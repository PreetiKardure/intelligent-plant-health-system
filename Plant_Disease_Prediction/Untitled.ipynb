```python
import pandas as pd

# ----- 1. Disease List -----
class_name = [
    'Apple___Apple_scab','Apple___Black_rot','Apple___Cedar_apple_rust',
    'Apple___healthy','Blueberry___healthy','Cherry_(including_sour)___Powdery_mildew',
    'Cherry_(including_sour)___healthy','Corn_(maize)___Cercospora_leaf_spot Gray_leaf_spot',
    'Corn_(maize)___Common_rust_','Corn_(maize)___Northern_Leaf_Blight','Corn_(maize)___healthy',
    'Grape___Black_rot','Grape___Esca_(Black_Measles)','Grape___Leaf_blight_(Isariopsis_Leaf_Spot)',
    'Grape___healthy','Orange___Haunglongbing_(Citrus_greening)','Peach___Bacterial_spot',
    'Peach___healthy','Pepper,_bell___Bacterial_spot','Pepper,_bell___healthy',
    'Potato___Early_blight','Potato___Late_blight','Potato___healthy','Raspberry___healthy',
    'Soybean___healthy','Squash___Powdery_mildew','Strawberry___Leaf_scorch','Strawberry___healthy',
    'Tomato___Bacterial_spot','Tomato___Early_blight','Tomato___Late_blight','Tomato___Leaf_Mold',
    'Tomato___Septoria_leaf_spot','Tomato___Spider_mites Two-spotted_spider_mite','Tomato___Target_Spot',
    'Tomato___Tomato_Yellow_Leaf_Curl_Virus','Tomato___Tomato_mosaic_virus','Tomato___healthy'
]

# ----- 2. Generate Disease Q&A -----
disease_questions = []
for disease in class_name:
    if "healthy" not in disease.lower():
        disease_questions.append({
            "Question": f"What is {disease}?",
            "Answer": f"{disease} detected. Please follow proper care and treatment instructions."
        })
        disease_questions.append({
            "Question": f"How to cure {disease}?",
            "Answer": f"To treat {disease}, remove affected leaves, avoid overwatering, and use fungicides if needed."
        })
        disease_questions.append({
            "Question": f"Can {disease} affect my yield?",
            "Answer": f"Yes, {disease} can reduce yield if untreated. Monitor and care accordingly."
        })

# ----- 3. Health & Trend Q&A -----
health_trend_questions = [
    {"Question": "How healthy is my plant?", "Answer": "Current health score is [health_score]/100."},
    {"Question": "What is my plant's health score?", "Answer": "Plant Health Score: [health_score]/100."},
    {"Question": "Is my plant improving?", "Answer": "Plant health trend is [trend]."},
    {"Question": "Is my plant declining?", "Answer": "Plant health trend is [trend]."},
    {"Question": "Can health score improve?", "Answer": "Yes, with proper care, [health_score] can increase."},
    {"Question": "What score is considered healthy?", "Answer": "80-100 is considered healthy."},
    {"Question": "What score is moderate stress?", "Answer": "40-59 indicates moderate stress."},
    {"Question": "What score is critical?", "Answer": "0-39 is critical; urgent care needed."},
]

# ----- 4. General Plant Care Q&A -----
general_care_questions = [
    {"Question": "How often should I water tomato plants?", "Answer": "Water 2-3 times a week or when soil is dry."},
    {"Question": "How much sunlight do apple trees need?", "Answer": "6-8 hours of sunlight daily."},
    {"Question": "Best fertilizer for potatoes?", "Answer": "Use nitrogen-rich fertilizer during growth stage."},
    {"Question": "How to prevent pest infestation?", "Answer": "Inspect leaves, use natural repellents, maintain hygiene."},
    {"Question": "How to prune grape plants?", "Answer": "Remove dead branches, prune after harvest season."},
    {"Question": "How to grow healthy strawberries?", "Answer": "Ensure proper sunlight, water, and remove diseased leaves."},
    {"Question": "How to care for blueberries?", "Answer": "Acidic soil, moderate water, avoid over-fertilization."},
    {"Question": "How to improve tomato yield?", "Answer": "Prune, proper watering, and control diseases."},
    {"Question": "How to prevent leaf scorch?", "Answer": "Avoid excessive sunlight and maintain adequate watering."},
    {"Question": "How to protect plants from frost?", "Answer": "Use covers, mulch, and avoid late planting."},
]

# ----- 5. Combine All Q&A -----
all_questions = disease_questions + health_trend_questions + general_care_questions

# ----- 6. Save to CSV -----
df = pd.DataFrame(all_questions)
df.to_csv("chatbot_data.csv", index=False)
print(f"CSV file created with {len(df)} questions!")
```
